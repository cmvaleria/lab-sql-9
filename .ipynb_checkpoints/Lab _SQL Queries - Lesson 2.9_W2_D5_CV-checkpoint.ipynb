#### 1) How many distinct (different) actors' last names are there?


```python
query = 'select count(distinct(last_name)) as count_last_name from sakila.actor'

count_distint_lastname = pd.read_sql_query(query, engine)
count_distint_lastname
```

#### 2) Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.


```python
query = '''select *
from sakila.rental;'''

rental = pd.read_sql_query(query, engine)
rental.head()


```

```python
rental['rental_date'] = pd.to_datetime(rental['rental_date'])
rental.head()
```

```python
def weekday(x):
    if x >=5:
        return 'weekend'
    else:
        return 'workday'

rental['day_type'] = rental['rental_date'].dt.dayofweek.apply(weekday)
## using lambda -- rental['day_type'] = rental['rental_date'].dt.dayofweek.apply(lambda x: 'weekend' if x>=5 else 'week')



# dt.dayofweek to extract the day of the week as an integer, where Monday is 0 and Sunday is 6.

rental
```

#### 4) Get 10 the longest films.


```python
query = 'select title, length \
FROM sakila.film \
ORDER by length desc \
LIMIT 10'

longuestmoevies = pd.read_sql_query(query, engine)
longuestmoevies
```

5) How many films include Behind the Scenes content?


```python
query = "select count(special_features) as movies_that_contain_bts \
from film \
where special_features = 'Behind the Scenes'"

behind_the_scenes = pd.read_sql_query(query, engine)
behind_the_scenes
```

6) Which kind of movies (rating) have a mean duration of more than two hours?


```python
query = 'select round(avg(length),2) as average_lengths, rating \
from sakila.film \
group by rating \
having round(avg(length),2) > 120;'

movies2h = pd.read_sql_query(query, engine)
movies2h
```

7) Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query = 'select title, length, dense_rank()  OVER (ORDER BY length DESC) as Rank_Ranking \
from film'

ranklength = pd.read_sql_query(query, engine)
ranklength
```
